```python
import yaml
import requests

url = yaml.load(open("../conf.yaml"), Loader=yaml.FullLoader)["url"]
```

```python
q = requests.get(f"{url}/health_check")
q.text
```

```python
data = {
            "query": "Газпром преобразован в акционерное общество",
            "search_engine": "bm25",
            "k": 5,
        }
requests.post(f"{url}/search", json=data).json()
```

```python
data = {
            "query": "Газпром преобразован в акционерное общество",
            "search_engine": "vector",
            "k": 5,
        }
requests.post(f"{url}/search", json=data).json()
```

```python
data = {
            "query": "Сделка Газпрома",
            "search_engine": "comb",
            "k": 5,
        }
requests.post(f"{url}/search", json=data).json()
```

```python
data = {
            "query": "Покупка газа",
            "search_engine": "comb",
            "k": 10,
        }
requests.post(f"{url}/search", json=data).json()
```

Из запросов выше видно, что bm25 более "буквальный", а vect менее. Лучше всего работает комбинированный поиск.
